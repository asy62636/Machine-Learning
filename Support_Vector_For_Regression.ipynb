What is Support Vector Regression?
Kernel Support vector machine and kernel support vector regression
In linear regression we used a single line
In SVR we have a tube, with vertical width epsilon $\epsilon$

Its called $\epsilon$-insensitive tube, it allows a little bit of buffer for the points about the axis of the tube. but there are also certain points that lie outside the tube, and we care about their error

They are denoted as $\xi$

$\xi$ $_{i}$ is above the tube and $\xi$ $_{j}^{*}$ below the tube
Here we do care about the errors 
The error here is given by:-

1/2 $\cdot$||$\omega$||$^{2}$ + C $\cdot$ $\sum_{i = 1}^{m}$($\xi$ $_{i}$ + $\xi$ $_{i}^{*}$)  --> Min

The svr accounts for some error in the method

Why is it called support vector regression?
Ans. Any point in the multidimensional space consisting of the data points is a vector and we care about them as vectors in our computation

In SVR model we need to apply feature scaling
There is no explicit equation depicting the relation between the dependent parameter and the features
We definitely have to apply feature scaling, since there are no coefficients that can compensate for high values of the features

```python
x = dataframe.iloc[:, 1:-1].values
y = dataframe.iloc[:, 2:].values
print(x,y)
```

Doing- Feature scaling:-
Feature scaling was not done on Y values in data preprocessing, since the Y values took value as 0 or 1. But,here the values take values from 45000 to 1000000

So now we need to apply feature scaling to be applied to the dependent variables as well. I we dont apply feature scaling, the values will not play out the way we need them to.
Make sure to give x and y as 2D arrays as input parameters to fit_transform and transform

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc2 = StandardScaler()
x = sc.fit_transform(x)
y = sc2.fit_transform(y)
```

Had to make 2 different standard scaler objects as x and y both have different standard deviation and and mean
